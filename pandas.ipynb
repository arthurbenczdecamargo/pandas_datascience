```python
data_apartments = pd.read_csv("data_apartments.csv", sep=";")
data_apartments.head(15)
```

```python
data_apartments["Price/month"] = (
    data_apartments["Valor"] + data_apartments["Condominio"]
)
data_apartments.head(15)
```

```python
data_apartments["Price/year"] = (
    data_apartments["Price/month"] * 12 + data_apartments["IPTU"]
)
data_apartments.head(15)
```

```python
data_apartments["Has_suite"] = data_apartments["Suites"].apply(
    lambda x: "Yes" if x > 0 else "No"
)
data_apartments.head(15)
```

```python
data_apartments["Description"] = (
    data_apartments["Quartos"].astype(str)
    + " room(s) apartment in "
    + data_apartments["Bairro"]
    + " with "
    + data_apartments["Vagas"].astype(str)
    + " garage space(s)."
)
data_apartments.head(10)
```

```python
data_apartments.to_csv('data_apartments_updated.csv', index=False, sep=';')
```
